```python
def clean_column_names(columns):
    new_columns = []
    for col in columns:
        col = col.strip()
        col = col.lower()
        col = re.sub(r"[^\w\s]"," ", col)
        col = "_".join(col.split())
        new_columns.append(col)
    return new_columns

df.columns = clean_column_names(df.columns)
print(df.columns)
df
```
